```python
# inport necessary modules

import pandas as pd
import numpy as np
import bqplot
import ipywidgets
```

```python
# read data and convert to pandas dataframe

df=pd.read_csv("F:/DEVANSHI/UIUC/sem2/IS590 Data Viz/week3/IL_Building_Inventory.csv", 
               na_values={'Year Acquired':0, 'Year Constructed':0, 'Square Footage':0})
```

```python
# group by the agencies and congress district

footage=df.groupby(['Agency Name','Congress Dist'])['Square Footage'].sum()
footage
```

```python
# unstack organizes the dataframe to look like an array which is useful for heat map

# data=pd.DataFrame(footage).reset_index()
data = footage.unstack(0)
#data
```

```python
# HeatMap 

x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()
c_sc = bqplot.ColorScale(colors=['white','yellow','red' ])

x_ax = bqplot.Axis(scale=x_sc, label='Governmental Department', tick_rotate=90, tick_style={'font_size':10})
y_ax = bqplot.Axis(scale=y_sc, label='Congress District', orientation='vertical', tick_values=np.arange(0,19))
c_ax = bqplot.ColorAxis(scale=c_sc, label='Square Footage', orientation='vertical', side='right')

heat_map = bqplot.GridHeatMap(color=data, 
                              row=data.index.values, 
                              column=data.columns.values, 
                              scales={'row':y_sc, 'column':x_sc, 'color':c_sc}, 
                              interactions={'click':'select'},
                             anchor_style={'fill':'green'}, 
                              selected_style={'opacity':1.0}, 
                              unselected_style={'opacity':0.6})

fig1 = bqplot.Figure(marks=[heat_map], axes=[x_ax, y_ax, c_ax])

#fig1
```

```python
# group by agency--congress-year for connecting the heat map to line plot

line_plot_data = df.groupby(['Agency Name','Congress Dist','Year Acquired'])['Square Footage'].sum()
#line_plot_data
```

```python
# Line Graph

x_sc_line = bqplot.OrdinalScale()
y_sc_line = bqplot.LogScale()

x_ax_line = bqplot.Axis(scale=x_sc_line, label='Year Acquired', tick_rotate=90)
y_ax_line = bqplot.Axis(scale=y_sc_line, label='Area (Square Footage)',orientation='vertical')

line_plot = bqplot.Lines(scales={'x':x_sc_line, 'y':y_sc_line})
fig2 = bqplot.Figure(marks=[line_plot], axes=[x_ax_line, y_ax_line])
# fig2
```

```python
# link the heat map to line plot by defining the change function

agency_dict = {}
congress_dict = {}
j=0
k=0
for agency, cong_dist in zip(data.columns.values, data.index.values):
    agency_dict[j] = agency
    j += 1
    congress_dict[k] = cong_dist
    k=k+1

def plot_change(change):
    y = int(heat_map.selected[0][0])
    x = int(heat_map.selected[0][1])
    line_plot.x = line_plot_data.loc[agency_dict[x]].loc[congress_dict[y]].index.values
    line_plot.y = line_plot_data.loc[agency_dict[x]].loc[congress_dict[y]].values

heat_map.observe(plot_change,'selected')


```

```python
# use ipywidgets to display both graphs together
# clicking on a particular cell in heat map shows the corresponding line graph for those parameters

display(ipywidgets.HBox([fig1,fig2]))
```
