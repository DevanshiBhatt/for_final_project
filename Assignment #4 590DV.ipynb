```python
# import necessary modules
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as colors
import matplotlib.cm as cm
#import ipywidgets
#import palettable
```

```python
# load the data
fn1 = "C:/Users/ANSHITA BHATT/Downloads/michigan_lld.flt"
michigan = np.fromfile(fn1, dtype='f4').reshape((5365, 4201))
```

```python
# Plot-1A
plt.style.use("classic")
plt.imshow(michigan)
plt.colorbar()
plt.clim(0,100)
```

### Explanation
Above image is the simplest visual of the lake Michigan and its surrounding areas. The color scale gives the depth above the 
sea level starting at 0. It is easy from this graph to identify and distinguish between the water and land regions.

However, this plot is not able to capture the details of given region. This is because the color limit does not cover the 
entire range of values for the image (minimum to maximum). It is evident from the plot that entire lake area is dark blue 
which means that the entire lake is at the same depth, which might not be the case. However, if the scale of color would go 
on the negative side, the map would better explain the depths at various points. Also, major porion of the plot is red so if 
the scale is increased a bit on the positive side too, then a better sense of the image can be made.

### Explanation

```python
# Plot-2
plt.imshow(michigan, cmap="gist_earth")
plt.clim(-250, 250)
plt.colorbar(extend = 'both')
```

### Explanation
In the above plot i have used the 'gist_earth' color map which looks more realistic as the given image is of a certain portion 
of Earth. This cmap is similar to the 'terrain' cmap with the difference in colors used across the range. The correct minimum 
and maximum values given to the scale correctly define the depth as in the second graph. The light green and yellow regions are areas with lesser altitudes while portions that are white are at a greater altitude (nearly 250). 

The change in depth at various points is seen easily and can be distinguished due to use of a different cmap. In addition to the
lake, now the land portions can also be identified based on depth. Also, the colors are not too bright so it would not strain the eyes even if the plot is looked at for a while.I have also made use of the arrowheads on both sides of the scale.

```python
# Plot-3
plt.imshow(michigan, cmap="coolwarm", norm = colors.SymLogNorm(10))
plt.clim(-250, 250)
plt.colorbar(extend = 'both')
```

### Explanation
In the above plot, I attempted to use a cmap from the 'Diverging' category of color maps. The good point about this cmap is that 
if the only concern of a viewer is to identify the boundary of the lake, then it will be very easy since the sudden change in 
color from blue to red is an obvious change. 

Also, I have demonstrated the use of log values for the depth range. While working with images and arrays which are spread across 
a large range of values, then a log scale would be advisable for use. This is because, if a normal scale is chosen for such cases,
it would use a large set of colors and the plot would end up looking bad visually. Also, it would be difficult to navigate
 through the large range of values and would unnecessarily capture very minute details which may not be required. The fact that
a viewer has to make out important findings just by visualizing the graph requires the plot to be visually good.
