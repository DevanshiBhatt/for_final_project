```python
import pandas as pd
import numpy as np
import matplotlib
import bqplot
import ipywidgets
```

```python
df=pd.read_csv("F:/DEVANSHI/UIUC/sem2/IS590 Data Viz/week3/IL_Building_Inventory.csv", na_values={'Year Acquired':0, 'Year Constructed':0, 'Square Footage':0})
#df
```

```python
# Scale for X and Y axis
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()

# Defining axis
ax_x = bqplot.Axis(scale=x_sc, label='Year Constructed')
ax_y = bqplot.Axis(scale=y_sc, label='Total Area (Sq.Ft.)', orientation='vertical')
```

```python
# Scatter plot
scatters = bqplot.Scatter(x=df["Year Constructed"], y=df.groupby("Year Constructed")["Square Footage"].sum(),
                          scales={'x': x_sc, 'y': y_sc})

# Making the plot interactive by using Fast Interval Selector
selector = bqplot.interacts.FastIntervalSelector(scale=y_sc, marks=[scatters])
scatters.unselected_style = {'opacity':0.8}
scatters.selected_style = {'fill': 'red', 'stroke':'black'}

#plot the figure
fig = bqplot.Figure(marks=[scatters], axes=[ax_x, ax_y], interaction=selector)
#display(fig)
```

```python
#selector.selected
```

```python
# Adding a widget to the plot
axis_dropdown = ipywidgets.Dropdown(description="Choose X-axis", options=['Year Constructed', 'Congress Dist','County', 'City'], 
                                    style={'description_width': 'initial'})
```

```python
# Implementing traitlets
def change_X(change):
    scatters.x = df[axis_dropdown.value]
    scatters.y = df.groupby(axis_dropdown.value)["Square Footage"].sum()
    ax_x.label = axis_dropdown.value

axis_dropdown.observe(change_X, 'value')
```

```python
# Combining the scatter plot and widget using VBox
ipywidgets.VBox([axis_dropdown, fig])
```

### Explanation
The Buildings dataset was used to plot different parameters using scatter plot. In order to visualize the total area (in square footage) as a function of different parameters like Congress District, Year and County, I have made use of the dropdown menu box to select one of these. Traitlets have been used to define the change function so that different parameters can be plotted in the same figure. 

The advantage of using widgets and traitlets in combination with bqplot is that it makes the visualization more interactive and allows the viewer to look at more information in the same block. The Fast Interval Selector is used here because it helps to highlight a certain portion of the plot which is useful to visualize in a better way. 

One shortcoming that I faced while implementing above components of visualization, was that the different scale for different parameters used on x-scale. If LinearScale is used then columns like City and County could not be scaled properly while in case of OrdinalScale parameters like Year Constructed and Congress District could not be scaled well.

### Question 2
Visualization #2: Work through the Wealth of Nations plot we demonstrated in class. Change at least 3 aspects of this visualization. Explain why you made the choices you did - how does this help visualize some aspect of this dataset that the original visualization does not?

```python
import pandas as pd
import bqplot
import ipywidgets

initial_year = 1800
```

```python
from wealth_of_nations import process_data, get_min_max, get_data

data = process_data('F:/DEVANSHI/UIUC/sem2/IS590 Data Viz/week6/nations.json')
#data
```

```python
income_min, income_max, life_exp_min, life_exp_max, pop_min, pop_max = get_min_max(data)
#income_min
```

```python
# tooltip
tt = bqplot.Tooltip(fields=['name', 'x', 'y', 'population'], 
                    labels=['Country Name', 
                            'Income per Capita', 'Life Expectancy', 'Population'])
```

```python
# Label
year_label = bqplot.Label(x=[0.5], y=[0.9], 
                   font_size=60, font_weight='bold', 
                   colors=['black'],
                   text=[str(initial_year)], enable_move=True)
```

```python
x_sc = bqplot.LogScale(min=income_min, max=income_max)
y_sc = bqplot.LinearScale(min=life_exp_min, max=life_exp_max)

c_sc = bqplot.OrdinalColorScale(domain=data['region'].unique().tolist(), 
                                colors=bqplot.CATEGORY10[:10])

size_sc = bqplot.LinearScale(min=pop_min, max=pop_max)
```

```python
ax_x = bqplot.Axis(label='Income per Capita', scale=x_sc, grid_lines='solid')
ax_y = bqplot.Axis(label='Life Expectancy', scale=y_sc, orientation='vertical', side='left', 
                   grid_lines='solid')
```

```python
cap_income, life_exp, pop = get_data(data,initial_year,initial_year)

wealth_scat = bqplot.Scatter(x=cap_income, y=life_exp, 
                             color=data['region'], size=pop,
                      names=data['name'], population=data['population'][0], display_names=False,
                      scales={'x': x_sc, 'y': y_sc, 'color': c_sc, 
                              'size': size_sc},
                      default_size=4112, tooltip=tt, 
                             animate=True, stroke='Black',
                      unhovered_style={'opacity': 0.2})
```

```python
fig = bqplot.Figure(marks=[wealth_scat, year_label], 
                    axes=[ax_x, ax_y],
             title='Health and Wealth of Nations')

# lets control the size in pixels too
fig.layout.min_width = '960px'
fig.layout.min_height = '640px'

display(fig)
```

### Change 1
In the first part of the visualization, I changed the position of y_label from bottom-right corner to top-center. Also, I changed the color of label from orange to black.

The y_label is an important part of the scatter plot as it is indicative of the data plotted for that year. So, i believe that its position should be at the center and top of the plot (where it is visible). Also, the font color was changed so that its contrast better matches with the colors of the plot. This position and color of the label makes it easy for the viewer of the plot to relate the date (year) with the associated scatter plot. 


### Change 2
The tooltip used in the original visualization gives additional information about data apart from the x and y components. The scatter plot is ranged in size based on population. However, this is not clear from the plot itself. A viewer would not be able to identify why the sizes are different unless he sees the code. 
Hence, I would like to include the population field as well in the tooltip that appears upon hovering over a specific point on the scatter plot.  

However, in doing so, the value of population could not be displayed. 

```python
nation_line = bqplot.Lines(x=data['income'][0], 
                           y=data['lifeExpectancy'][0], 
                           colors=['Gray'],
                       scales={'x': x_sc, 'y': y_sc}, visible=False)
```

```python
def hover_changed(change):
    if change.new is not None:
        nation_line.x = data['income'][change.new + 1]
        nation_line.y = data['lifeExpectancy'][change.new + 1]
        nation_line.visible = True
    else:
        nation_line.visible = False
        
wealth_scat.observe(hover_changed, 'hovered_point')
```

```python
fig2 = bqplot.Figure(marks=[wealth_scat, year_label, nation_line], 
                    axes=[ax_x, ax_y],
             title='Health and Wealth of Nations')

fig2.layout.min_width = '960px'
fig2.layout.min_height = '640px'

#display(fig2)
```

```python
# setting time-interval for play button widget
time_interval = 10
```

```python
# drop-down widget
year_dropdown = ipywidgets.Dropdown(description="Choose Year", options=np.arange(1800, 2008, 1), 
                                    style={'description_width': 'initial'})

#play_button = ipywidgets.Play(min=1800, max=2008, interval=time_interval)
```

```python
# Implementing traitlets
def year_changed(change):
    wealth_scat.x, wealth_scat.y, wealth_scat.size = get_data(data, year_dropdown.value,initial_year)
    year_label.text = [str(year_dropdown.value)]

year_dropdown.observe(year_changed, 'value')
#play_button.observe(year_changed, 'value')
```

```python
fig3 = bqplot.Figure(marks=[wealth_scat, year_label, nation_line], 
                    axes=[ax_x, ax_y],
             title='Health and Wealth of Nations', animation_duration=time_interval)

fig3.layout.min_width = '960px'
fig3.layout.min_height = '640px'
```

```python
ipywidgets.VBox([year_dropdown, fig3])
```

```python
# a separate figure to use the play button for showing the general trend of change in total area along the years

play_button = ipywidgets.Play(min=1800, max=2008, interval=time_interval)

def year_changed(change):
    wealth_scat.x, wealth_scat.y, wealth_scat.size = get_data(data, play_button.value, initial_year)
    year_label.text = [str(play_button.value)]

play_button.observe(year_changed, 'value')
```

```python
fig4 = bqplot.Figure(marks=[wealth_scat, year_label, nation_line], 
                    axes=[ax_x, ax_y],
             title='Health and Wealth of Nations', animation_duration=time_interval)

# lets control the size in pixels too
fig4.layout.min_width = '960px'
fig4.layout.min_height = '640px'
```

```python
ipywidgets.VBox([ipywidgets.HBox([play_button]), fig4])
```

### Change 3
The original visualization uses play button and integer slider for year. However, in my opinion, a drop-down menu suits better for choosing a particular year. The integer slider is generally used for integer values and not timelines (here, dates in terms of years). Also, the slider and play button eventually show the same information so that does not really prove any significant purpose in the same figure. So, in case the purpose is to only depict the general trend, then the slider has no use. So I chose to use a drop-down box for year selection.

The drawback that I faced while implementing this was that the drop-down boz cannot be synced with any other widget which is why i had to create 2 separate figures- one that uses a drop-down and one that uses a play button. The play button is really helpful in showing a general trend in the data over a period. So by setting an appropriate value for time interval, the visual can be made really interactive.
